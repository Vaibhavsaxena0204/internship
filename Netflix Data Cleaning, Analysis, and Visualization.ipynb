# Netflix Data Cleaning, Analysis, and Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "netflix1.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic information
print("Dataset Information:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Data Cleaning
print("\nHandling missing values:")
df.fillna("Unknown", inplace=True)  # Fill missing values with 'Unknown'
```

```python
# Convert date_added to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Data Analysis
print("\nUnique content types:")
print(df['type'].value_counts())
```

```python
# Top 10 countries with the most content
country_counts = df['country'].value_counts().head(10)
```

# Visualization

```python
# Visualization
plt.figure(figsize=(10, 5))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='coolwarm')
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.title("Top 10 Countries with Most Netflix Content")
plt.show()
```

```python
# Distribution of Content by Rating
plt.figure(figsize=(10, 5))
sns.countplot(y=df['rating'], order=df['rating'].value_counts().index, palette='viridis')
plt.xlabel("Count")
plt.ylabel("Rating")
plt.title("Distribution of Netflix Content by Rating")
plt.show()
```

```python
# Count of Movies vs TV Shows
plt.figure(figsize=(7, 5))
sns.countplot(x=df['type'], palette='Set2')
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.title("Count of Movies vs TV Shows")
plt.show()
```

```python
# Trends of content release over the years
plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=30, kde=True, color='blue')
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Trend of Netflix Content Release Over the Years")
plt.show()
```
